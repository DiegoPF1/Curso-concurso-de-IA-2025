```python
import cv2 #Importa Open CV 2
import numpy as np  #Importa Numpy
import matplotlib.pyplot as plt #Para visualizar. 
                                #En local usar directamente 
                                #imshow
```

```python
ruta_imagen = 'imagenes/r2d2_urdf.png' #ruta a la imagen
```

```python
#carga la imagen (BGR)
imagen_bgr = cv2.imread(ruta_imagen)   
#convierte la imagen a espacio de color HSV
imagen_hsv = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2HSV)
```

```python
# Definir el rango para el color magenta en HSV
# El color magenta se encuentra aproximadamente en los rangos:
# H: 140-170 (en OpenCV va de 0 a 179), S: 50-255, V: 50-255
rango_bajo = np.array([100, 100, 50])   # puedes ajustar estos valores
rango_alto = np.array([130, 255, 255])
```

```python
# Crear una máscara donde se conserve solo el magenta
mascara_magenta = cv2.inRange(imagen_hsv, rango_bajo, rango_alto)
# Aplicar la máscara sobre la imagen original
resultado_magenta = cv2.bitwise_and(imagen_bgr, imagen_bgr, mask=mascara_magenta)
```

```python
# Convertir imágenes BGR a RGB para mostrar con matplotlib
imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)
resultado_rgb = cv2.cvtColor(resultado_magenta, cv2.COLOR_BGR2RGB)
```

```python
# Mostrar la imagen original y la imagen filtrada (solo magenta)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagen_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(resultado_rgb)
plt.title('Solo color magenta')
plt.axis('off')

plt.tight_layout()
plt.show()
```
